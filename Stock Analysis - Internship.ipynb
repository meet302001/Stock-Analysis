```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import datetime as dt
```

```python
data = pd.read_excel("sample_data.xls",index_col=[0])
```

```python
data.head()
```

```python
data.shape 
```

```python
data['Dates'] = pd.to_datetime(data['date']).dt.date
data['Time'] = pd.to_datetime(data['date']).dt.time
data.drop(['date','Time'],axis=1,inplace=True)
data.set_index('Dates')
```

```python
data.info()
```

```python
def set_fig_size():    
    f = plt.figure()
    f.set_figwidth(16)
    f.set_figheight(8)
```

```python

set_fig_size()

plt.plot(data['Dates'],data['close'],label = 'Close')
plt.plot(data['Dates'],data['open'],label = 'Open')
plt.plot(data['Dates'],data['high'],label = 'High')
plt.plot(data['Dates'],data['low'],label = 'Low')
plt.xlabel('Dates')
plt.grid()
plt.legend()
```

# Simple Moving Average (SMA)


```python
# 20 days moving average
data['20_SMA'] = data['close'].rolling(window = 20, min_periods = 1).mean()
```

```python
set_fig_size()
plt.plot(data['Dates'],data['close'],label='Close',markersize = 40)
plt.plot(data['Dates'],data['20_SMA'],label='SMA_close')
plt.xlabel('Dates')
plt.grid()
plt.legend()
```

# Exponential Moving Average (EMA)


```python
data['10_EMA'] = data['close'].ewm(span=10, adjust=False).mean()
```

```python
set_fig_size()
plt.plot(data['Dates'],data['close'],label='Close')
plt.plot(data['Dates'],data['10_EMA'],label='EMA_10')
plt.xlabel('Dates')
plt.grid()
plt.legend()
```

# Solving Tasks

```python
df = pd.read_excel("sample_data.xls",index_col=[0])
df.head()
```

```python
df['Date'] = pd.to_datetime(df['date']).dt.date
df['Time'] = pd.to_datetime(df['date']).dt.time
df.head(30)
```

```python
df['Date'] = df['Date'].astype(str)
df['Time'] = df['Time'].astype(str)

date = list(df['Date'])
buy = []
sell = []

for i in range(len(df['Time'])):
    if df['Time'][i] == "09:15:00":
        buy.append(100 * df['open'][i])
    if df['Time'][i] == "15:00:00":
        sell.append(100 * df['close'][i])

        
print(buy)
print()
print("~"*90)
print()
print(sell)

```

```python
# creating a data frame

output = pd.DataFrame(list(zip(date, buy, sell)),
               columns =['Date', 'Buy','Sell'])
```

```python
max_profit_loss = [(buy[i]-sell[i]) for i in range(len(buy))]
print(" Max Profit is :-  ", max(max_profit_loss))
print(" Max Loss is :-  ", abs(min(max_profit_loss)))
```
